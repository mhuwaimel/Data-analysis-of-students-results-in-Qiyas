```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(rc={'figure.figsize': [8, 8]}, font_scale=1.2)
```

```python
DataFrame=pd.read_csv("All.csv")
DataFrame.info()
```

## Clean Data

```python
DataFrame['Name'] = DataFrame['Name'].astype(pd.StringDtype())
DataFrame['City'] = DataFrame['City'].astype(pd.StringDtype())
DataFrame['Area'] = DataFrame['Area'].astype(pd.StringDtype())
DataFrame['Department'] = DataFrame['Department'].astype(pd.StringDtype())
DataFrame['Office'] = DataFrame['Office'].astype(pd.StringDtype())
DataFrame['Authority'] = DataFrame['Authority'].astype(pd.StringDtype())
DataFrame['EducationType'] = DataFrame['EducationType'].astype(pd.StringDtype())
DataFrame['ExamType'] = DataFrame['ExamType'].astype(pd.StringDtype())
DataFrame['MarkType'] = DataFrame['MarkType'].astype(pd.StringDtype())
DataFrame['ExamSpecial'] = DataFrame['ExamSpecial'].astype(pd.StringDtype())
DataFrame["KsaTarteeb42"]=DataFrame["KsaTarteeb42"].astype("Int64")
DataFrame["StdCount42"]=DataFrame["StdCount42"].astype("Int64")
DataFrame["KsaTarteeb41"]=DataFrame["KsaTarteeb41"].astype("Int64")
DataFrame["StdCount41"]=DataFrame["StdCount41"].astype("Int64")
DataFrame.head(15)
```

# تجزئة البيانات الى قدرات وتحصيلي 

```python
Qudurat_Data=DataFrame.query('ExamType=="قدرات" & Year==1444')
Tahsili_Data=DataFrame.query('ExamType=="تحصيلي" & Year==1444')
print(Qudurat_Data.shape)
print(Tahsili_Data.shape)
```

```python
# حساب متوسط درجات القدرات حسب المنطقة
area_avg_q = Qudurat_Data.groupby('Area')['Average'].mean().sort_values(ascending=False)
print("\nمتوسط القدرات حسب المنطقة: ")
print(area_avg_q)
```

```python
# حساب متوسط درجات التحصيلي حسب المنطقة
area_avg_t = Tahsili_Data.groupby('Area')['Average'].mean().sort_values(ascending=False)
print("\nمتوسط التحصيلي حسب المنطقة: ")
print(area_avg_t)
```

```python
# أفضل 15 مدرسة حسب معدل القدرات
top_schools_by_qudurat = Qudurat_Data.nlargest(15, 'Average')[['Name',  'Area', 'Average']]
print(top_schools_by_qudurat)
```

```python
# أفضل 15 مدرسة حسب معدل القدرات
top_schools_by_tahsili = Tahsili_Data.nlargest(15, 'Average')[['Name',  'Area', 'Average']]
print(top_schools_by_tahsili)
```

```python
# رسم توزيع معدلات القدرات
import arabic_reshaper
from bidi.algorithm import get_display
def reshape_arabic(txt):
    return get_display(arabic_reshaper.reshape(txt))


plt.figure(figsize=(10, 6))
sns.histplot(Qudurat_Data['Average'], bins=20, kde=True,color="blue")
plt.title(reshape_arabic('توزيع معدلات القدرات'))
plt.xlabel(reshape_arabic('معدل القدرات'))
plt.ylabel(reshape_arabic('عدد المدارس'))
plt.show()
```

```python
# تجميع البيانات حسب نوع المدرسة
management_analysis = Qudurat_Data.groupby('Authority')['Average'].mean()
print(management_analysis)
```

```python
# رسم متوسط معدلات القدرات حسب نوع المدرسة
plt.figure(figsize=(8, 5))
sns.barplot(x=management_analysis.index, y=management_analysis.values,formatter=reshape_arabic, color="blue")
plt.title(reshape_arabic('مقارنة بين نوع المدارس'))
plt.xlabel(reshape_arabic('نوع المدرسة'))
plt.ylabel(reshape_arabic('متوسط معدل القدرات'))
plt.show()
```

```python
# تجميع البيانات حسب المنطقة وحساب المتوسط لكل إدارة تعليم
qudurat_region_analysis = Qudurat_Data.groupby('Department')['Average'].mean().sort_values(ascending=False)
```

```python
# رسم أفضل المناطق حسب معدلات القدرات
plt.figure(figsize=(12, 6))
sns.barplot(x=qudurat_region_analysis.index, y=qudurat_region_analysis.values,formatter=reshape_arabic,color="navy")
plt.title(reshape_arabic('أفضل الإدارات حسب معدلات القدرات'))
plt.xlabel(reshape_arabic('الإدارة'))
plt.ylabel(reshape_arabic('متوسط معدل القدرات'))
plt.xticks(rotation=90)
plt.show()
```

```python
# رسم أفضل المناطق حسب معدلات التحصيلي
tahsili_region_analysis=Tahsili_Data.groupby('Department')['Average'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=tahsili_region_analysis.index, y=tahsili_region_analysis.values,formatter=reshape_arabic,color="blue")
plt.title(reshape_arabic('أفضل الإدارات حسب معدلات التحصيلي'))
plt.xlabel(reshape_arabic('الإدارة'))
plt.ylabel(reshape_arabic('متوسط معدل التحصيلي'))
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# التنبؤ لكل Department
dept_year_avg = DataFrame.groupby(['Department', 'Year'])['Average'].mean().reset_index()

future_years = pd.DataFrame({'Year': range(1445, 1450)})

def predict_future_averages(dept_data, future_years):
    X = dept_data['Year'].values.reshape(-1, 1)
    y = dept_data['Average'].values
    model = LinearRegression()
    model.fit(X, y)
    future_X = future_years['Year'].values.reshape(-1, 1)
    predictions = model.predict(future_X)
    return predictions

predictions_dict = {}
for dept in dept_year_avg['Department'].unique():
    dept_data = dept_year_avg[dept_year_avg['Department'] == dept]
    if len(dept_data) >= 3:
        predictions = predict_future_averages(dept_data, future_years)
        predictions_dict[dept] = predictions
    else:
        predictions_dict[dept] = "بيانات غير كافية"
print("\nالتنبؤات للخمس سنوات القادمة:")
for dept, preds in predictions_dict.items():
    print(f"{dept}: {preds}")
```
